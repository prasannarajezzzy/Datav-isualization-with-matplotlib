```python
#subplot
plt.figure(1)
x=np.linspace(20,30,40)
y=np.log(x)


plt.subplot(221)
plt.plot(x,y)
plt.title("sady")


plt.subplot(222)
plt.plot(x,y**9)
plt.title("anj")

plt.figure(2)


plt.subplot(221)
plt.plot(x,y**2)
plt.title("prasabna")


plt.subplot(222)
plt.plot(x**2,y*2)
plt.title("love")
plt.show()


```
